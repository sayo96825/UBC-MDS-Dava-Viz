```python
import altair as alt
import pandas as pd
alt.themes.enable('fivethirtyeight')
```

```python
gm_url = 'https://raw.githubusercontent.com/UBC-MDS/exploratory-data-viz/main/chapters/en/slides/module2/data/world-data-gapminder.csv'
gm_df = pd.read_csv(gm_url,parse_dates=['year'])
gm_df.head()
```

```python
gm_df.info()
```

```python
movies_df = pd.read_csv('https://raw.githubusercontent.com/UBC-MDS/exploratory-data-viz/main/chapters/en/slides/module4/data/movies-extended.csv')
```

```python
sort_order = ['Concert/Performance', 'Musical', 'Documentary', 'Adventure', 
              'Comedy', 'Romantic Comedy', 'Drama',  'Action']
alt.Chart(movies_df[movies_df['MPAA Rating'].isin(['G', 'PG'])]).mark_bar().encode(
    alt.X('count()', stack='normalize', title='Proportion of movies'),
    alt.Y('Major Genre', sort=sort_order),
    alt.Color('MPAA Rating'),
    alt.Order('MPAA Rating'))
```

```python
import altair as alt

region_order = ['Asia', 'Africa','Americas','Europe','Oceania']

alt.Chart(gm_url).mark_area(opacity= 0.7).encode(
    alt.X('year:T'),
    alt.Y('sum(population):Q',title = "Cumulative world population"),
    alt.Color('region:N', sort = 'descending')).properties(title='World population by Region')
```

```python
plot1 = alt.Chart(gm_df2018).mark_bar().encode(
    alt.X('sum(population)', title='Population'),
    alt.Y('region', sort = 'x'))

popu_order = ['Oceania','Europe','Americas','Africa','Asia']

plot2 = alt.Chart(gm_df2018).mark_bar().encode(
    alt.X('count()'),
    alt.Y('region', sort = popu_order))

plot1 | plot2
```

```python
alt.Chart(gm_df2018).mark_bar().encode(
    alt.X('life_expectancy', bin=True),
    alt.Y('count()'))
```

```python
alt.Chart(gm_df2018).mark_bar().encode(
    alt.X('life_expectancy', bin=alt.Bin(maxbins=30), title='Life expectancy'),
    alt.Y('count()'),
    alt.Color('region')
).properties(width=200, height=150
).facet('region', columns = 2)
```

```python
alt.Chart(gm_df2018[gm_df2018['income_group'].isin(['High', 'Low'])]).mark_bar().encode(
    alt.X('life_expectancy', bin=alt.Bin(maxbins=30)),
    alt.Y('count()'),
    alt.Color('region')
).properties(width=200, height=140
).facet(column='region', row='income_group')
```

```python
income_order = ['High', 'Upper middle', 'Lower middle', 'Low']
alt.Chart(gm_df2018[gm_df2018['region'].isin(['Africa'])]).mark_point().encode(
    alt.X('children_per_woman'),
    alt.Y('child_mortality'),
    alt.Color('income_group', sort=income_order),
).properties(width=200, height=140
).facet('sub_region')
```

```python
alt.Chart(gm_df2018[gm_df2018['region'] == 'Africa']).mark_bar().encode(
    alt.X('count()', title='Number of countries'),
    alt.Y('income_group', sort=income_order),
    alt.Color('income_group', sort=income_order, legend=None)
).properties(width=200
).facet('sub_region')
```
